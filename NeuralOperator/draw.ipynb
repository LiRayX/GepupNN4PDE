```python
import matplotlib.pyplot as plt
import numpy as np
```

```python

# 
steps = []
train_losses = []
test_losses = []
test_metrics = []

try:
    # 
    with open('output_x_new.txt', 'r') as file:
        lines = file.readlines()
        # 
        for line in lines[1:]:
            parts = line.split()
            if len(parts) == 4:
                steps.append(int(parts[0]))
                train_losses.append(float(parts[1].strip('[]')))
                test_losses.append(float(parts[2].strip('[]')))
                test_metrics.append(float(parts[3].strip('[]')))
            else:
                print(f"Skipping line due to unexpected format: {line}")

    #  numpy 
    steps = np.array(steps)
    train_losses = np.array(train_losses)
    test_losses = np.array(test_losses)
    test_metrics = np.array(test_metrics)

    # 
    if len(steps) == 0 or len(train_losses) == 0 or len(test_losses) == 0 or len(test_metrics) == 0:
        raise ValueError("No data was read from the file.")
    
    # 
    fig, axs = plt.subplots(2, 3, figsize=(18, 10))

    # 
    axs[0, 0].scatter(steps[:3], train_losses[:3], label='Train loss', color='blue')
    for i in range(3):
        axs[0,0].text(steps[i], train_losses[i], f'{train_losses[i]:.6e}', fontsize=9, ha='left')
    axs[0, 0].set_xlabel('Step')
    axs[0, 0].set_ylabel('Train loss')
    axs[0, 0].set_title('Train Loss (First 2000 Steps)')
    axs[0, 0].grid(True)

    axs[0, 1].scatter(steps[:3], test_losses[:3], label='Test loss', color='green')
    for i in range(3):
        axs[0,1].text(steps[i], test_losses[i], f'{test_losses[i]:.6e}', fontsize=9, ha='left')
    axs[0, 1].set_xlabel('Step')
    axs[0, 1].set_ylabel('Test loss')
    axs[0, 1].set_title('Test Loss (First 2000 Steps)')
    axs[0, 1].grid(True)

    axs[0, 2].scatter(steps[:3], test_metrics[:3], label='Test metric', color='red')
    for i in range(3):
        axs[0,2].text(steps[i], test_metrics[i], f'{test_metrics[i]:.6e}', fontsize=9, ha='left')
    axs[0, 2].set_xlabel('Step')
    axs[0, 2].set_ylabel('Test metric')
    axs[0, 2].set_title('Test Metric (First 2000 Steps)')
    axs[0, 2].grid(True)

    # 
    axs[1, 0].plot(steps[3:], train_losses[3:], label='Train loss', color='blue')
    axs[1, 0].set_xlabel('Step')
    axs[1, 0].set_ylabel('Train loss')
    axs[1, 0].set_title('Train Loss (Remaining Steps)')
    axs[1, 0].grid(True)

    axs[1, 1].plot(steps[3:], test_losses[3:], label='Test loss', color='green')
    axs[1, 1].set_xlabel('Step')
    axs[1, 1].set_ylabel('Test loss')
    axs[1, 1].set_title('Test Loss (Remaining Steps)')
    axs[1, 1].grid(True)

    axs[1, 2].plot(steps[3:], test_metrics[3:], label='Test metric', color='red')
    axs[1, 2].set_xlabel('Step')
    axs[1, 2].set_ylabel('Test metric')
    axs[1, 2].set_title('Test Metric (Remaining Steps)')
    axs[1, 2].grid(True)

    # 
    for ax in axs.flat:
        ax.legend()

    # 
    plt.tight_layout()
    plt.show()

except FileNotFoundError:
    print("The file 'data.txt' was not found.")
except ValueError as ve:
    print(f"Value error: {ve}")
except Exception as e:
    print(f"An unexpected error occurred: {e}")
```
